```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.kap.org.tr/tr/bist-sirketler")
response
```

```python
bs = BeautifulSoup(response.content, "lxml")
```

```python
bs.prettify()
```

```python
bist = bs.find_all("div", attrs={"class":"comp-cell _04 vtable"})
bist
```

```python
bistKod = bs.find_all("div", attrs={"class":"comp-cell _04 vtable"})
kapUrl = "https://www.kap.org.tr"
kodList, hisseUrlList = list(), list()
for b in bistKod:
#     print(b.find("a", attrs={"class":"vcell"}))
    hisseLink = kapUrl + b.a.get("href")
    hisseKod = b.text
    kodList.append(hisseKod.strip())
    hisseUrlList.append(hisseLink.strip())
len(kodList), len(hisseUrlList)
```

```python
unvanList = []
bistUnvan = bs.find_all("div", attrs={"class":"comp-cell _14 vtable"})
for u in bistUnvan:
    unvan = u.text
    unvanList.append(unvan.strip())
len(unvanList)
```

```python
cityList = []
cities = bs.find_all("div", attrs={"class":"comp-cell _12 vtable"})
for c in cities:
    city = c.text
    cityList.append(city.strip())
len(cityList)
```

```python
bagimsizDenetimList = []
bagimsizDenetim = bs.find_all("div", attrs={"class":"comp-cell _11 vtable"})
for b in bagimsizDenetim:
    bd = b.text
    bagimsizDenetimList.append(bd.strip())
len(bagimsizDenetimList)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({"Kod":kodList,
             "Unvan":unvanList,
             "City":cityList,
             "BagimsizDenetim":bagimsizDenetimList,
             "Url":hisseUrlList})
data
```
